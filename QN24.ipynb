```python
import shap
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('breast-cancer.csv')
```

```python
df.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(['id','diagnosis'],axis=1),df['diagnosis'],test_size=0.24,random_state=101)
```

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
explainer=shap.TreeExplainer(model)
```

```python
shap.summary_plot(explainer.shap_values(X_test),X_test)
```

```python
shap_values=explainer.shap_values(X_test)
```

```python
shap.initjs()
shap.plots.force(explainer.expected_value[1],shap_values[1])
```
