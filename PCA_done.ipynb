Qn. Implement dimensionality reduction using **PCA**

1. Use the Iris Dataset present in the scikit-learn library
2. implement PCA to reduce dimension into 2 dimensions [PC1, PC2]
4. Display new reduced dimension values
5. Plot appropriate graph/ label w.r.t. principal component.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
```

```python
# Loading the dataset
iris = datasets.load_iris()

X = iris.data
y = iris.target
print("X:",X[0])
target_names = iris.target_names


# Scaling the data
scaler = MinMaxScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
# Define a function for plotting graphs
def plot3clusters(X, title, vtitle):
  colors=['r','b','y']
  plt.figure()
  for color,i,name in zip(colors,[0,1,2],target_names):
    plt.scatter(X[y==i,0],X[y==i,1],color=color,label=name)
    plt.xlabel(vtitle+'1')
    plt.ylabel(vtitle+'2')
    plt.show()
# plot3clusters(X_scaled,'AE','AE')


# Implement and visualize PCA
pca = PCA(n_components=2)
transformedData = pca.fit_transform(X_scaled)
print('PCA transformed',transformedData[0])
plot3clusters(transformedData,'PCA','PC')

#************************#
```
